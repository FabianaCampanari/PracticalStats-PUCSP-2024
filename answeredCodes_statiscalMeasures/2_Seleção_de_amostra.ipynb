```python
#verifica as classes da variável e sua proporção
perc_est_civ=df["estado_civil"].value_counts(normalize = True)

perc_est_civ
```

```python
# Executa a amostra estratificada igualitária por estado civil
amostra_est_igual=df.groupby(['estado_civil'], group_keys=False).apply(lambda x: x.sample(n=10,replace=False,random_state=2903))

amostra_est_igual
```

```python
#definir total desejado
N = 20

# Executa a amostra estratificada proporcional por estado civil
amostra_est_prop = df.groupby(['estado_civil'], group_keys=False).apply(
    lambda x: x.sample(int(np.rint(N * len(x) / len(df))), random_state=2903)  # Cálculo proporcional da amostra
).sample(frac=1, random_state=2903).reset_index(drop=True)  # Embaralhar a amostra

amostra_est_prop
```

```python

#verifica a proporção de cada estado civil
perc_est_civ=amostra_est_prop["estado_civil"].value_counts(normalize = True)
perc_est_civ
```

```python
# Executa a amostra estratificada igualitária por regiao de procedencia
amostra_est_igual=df.groupby(['reg_proc'], group_keys=False).apply(lambda x: x.sample(n=10,replace=False,random_state=2903))

amostra_est_igual
```

```python
# Executa a amostra estratificada igualitária por grau de instrução
amostra_est_igual=df.groupby(['grau_instrucao'], group_keys=False).apply(lambda x: x.sample(n=10,replace=False,random_state=2903))

amostra_est_igual
```

```python
# Executa a amostra estratificada igualitária por reg_proc e grau de instrucao
amostra_est_igual=df.groupby(['reg_proc','grau_instrucao'], group_keys=False).apply(lambda x: x.sample(n=10,replace=False,random_state=2903))
amostra_est_igual
```


# min(len(x), 10): Garante que, se um estrato tiver menos de 10 elementos, ele não tentará amostrar mais do que o disponível.

# reset_index(drop=True): Reseta o índice do DataFrame resultante para evitar índices duplicados.

# Este código cria uma amostra estratificada igualitária por grau de instrução e região de procedência, considerando até 10 casos por estrato, e exibe a amostra resultante.

# 👇 To save the stratified sample to a new Excel file, you can use the to_excel method provided by pandas. Here is the complete code including the steps to save the sample:



```python
# Importar bibliotecas pandas e numpy
import pandas as pd
import numpy as np

# Definir o caminho do arquivo
caminho = '/Users/fabicampanari/Desktop/_8-Prova Matematematica/1-statiscalMeasures_ Hypothesis Testing II./Dataset/cadastro_funcionarios.xlsx'

# Ler o arquivo Excel em um DataFrame
df = pd.read_excel(caminho)

# Criar uma amostra estratificada igualitária por grau de instrução e região de procedência
amostra_est_igual = df.groupby(['grau_instrucao', 'reg_proc'], group_keys=False).apply(lambda x: x.sample(n=min(len(x), 10), replace=False, random_state=2903)).reset_index(drop=True)

# Exibir a amostra resultante
print(amostra_est_igual)

# Salvar a amostra estratificada em um novo arquivo Excel
caminho_saida = '/Users/fabicampanari/Desktop/amostra_estratificada.xlsx'
amostra_est_igual.to_excel(caminho_saida, index=False)

print(f"A amostra estratificada foi salva em {caminho_saida}")
```
